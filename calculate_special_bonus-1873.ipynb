```python
import pandas as pd
```

```python
data = [[2, 'Meir', 3000], [3, 'Michael', 3800], [7, 'Addilyn', 7400], [8, 'Juan', 6100], [9, 'Kannon', 7700]]
employees = pd.DataFrame(data, columns=['employee_id', 'name', 'salary']).astype({'employee_id':'int64', 'name':'object', 'salary':'int64'})
```

```python
resultado = employees[(employees['employee_id'] % 2 != 0) & (employees['name'].str[0] != 'M')]
resultado.loc[:, 'bonus'] = resultado['salary']
resultado = resultado[['employee_id', 'bonus']]
resultado
```

```python
emp_bonus = employees.merge(right = resultado, how= 'left', left_on = 'employee_id', right_on = 'employee_id', suffixes = ['_emp', '_bonus'])
emp_bonus.sort_values(by='employee_id', ascending=True)[['employee_id', 'bonus']].fillna(0)
```
