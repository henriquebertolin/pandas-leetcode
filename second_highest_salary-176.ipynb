```python
import pandas as pd
data = [[1, 100], [2, 200], [3, 300]]
employee = pd.DataFrame(data, columns=['id', 'salary']).astype({'id':'int64', 'salary':'int64'})
```

```python
employee['rank'] = employee['salary'].rank(method = 'dense', ascending= False)
employee
```

```python
secondHighestSalary = employee[employee['rank']==2]
secondHighestSalary
```

```python
second = pd.DataFrame({'SecondHighestSalary':[None if len(secondHighestSalary) == 0 else secondHighestSalary['salary'].iloc[0]]})
second
#return second
```
