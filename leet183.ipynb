```python
import pandas as pd

data = [[1, 'Joe'], [2, 'Henry'], [3, 'Sam'], [4, 'Max']]
customers = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
data = [[1, 3], [2, 1]]
orders = pd.DataFrame(data, columns=['id', 'customerId']).astype({'id':'Int64', 'customerId':'Int64'})
orders
```

```python
cust_ord = customers.merge(right = orders, how ='left', left_on = 'id', right_on = 'customerId', suffixes = ['_cust','_ord'])
cust_ord
```

```python
cust_ord.rename(columns={
    'name' : 'Customers'
}, inplace = True)
resultado = cust_ord[cust_ord['customerId'].isna()]
resultado[['Customers']]
```
