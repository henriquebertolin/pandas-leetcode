```python
import pandas as pd
```

```python
data = [[1, 3.5], [2, 3.65], [3, 4.0], [4, 3.85], [5, 4.0], [6, 3.65]]
scores = pd.DataFrame(data, columns=['id', 'score']).astype({'id':'Int64', 'score':'Float64'})
```

```python
scores
```

```python
sortedScores = scores.sort_values(by='score', ascending= False)
sortedScores
```

```python
sortedScores['rank'] = sortedScores[['score']].rank(method='dense', ascending=False)
sortedScores[['score', 'rank']]
```
